```python
df=pd.read_csv("C:/Users/Eduardo CV/Desktop/ejericios python/avocado/avocado.csv")
df.head(7)
```

```python
# obtener datos SÖLAMENTE de Atlanta:

atlanta=df[df["region"]=="Atlanta"]

atlanta.head(6)
```

```python
# extraer columnas de la tabla "Atlanta"
precio=atlanta["AveragePrice"]
precioPromediado=precio.rolling(30, min_periods=1).mean()# tomca muestra de (cada) 30 datos y saca la media.. así hasta completar datos
volumen=atlanta["Total Volume"]

bolsasAguacate=atlanta["Total Bags"]
sAguacate=atlanta["Small Bags"]
lAguacate=atlanta["Large Bags"]
xlAguacate=atlanta["XLarge Bags"]
```

```python
# ggraficar dos imagenes en una gráfica
    #grafíca 1
plt.subplot(221) # 221, indica 2 filas, 2 columnas y 1 orden
plt.title("Precio del aguacate") #título
plt.plot(precio, label="Precio", color="green") #graficar precio, color verde
plt.plot(precioPromediado, label="Precio Promediado", color="orange")
plt.legend() #para generar leyenda
    #grafica 2
plt.subplot(222) # fila 2, columna 2, 2 orden
plt.title("Volumen de aguacates")
plt.plot(volumen, label="Volumen total", color="red")
plt.legend()
    #grafica 3
plt.subplot(223) # fila 2, columna 2, 3 posicion
plt.title("Bolsa total de aguacates")
plt.plot(bolsasAguacate, label="Bolsa total", color="blue")
plt.legend()
    #grafica 4
plt.subplot(224) # fila 2, columna 2, 3 posicion
plt.title("Bolsa por tamaño")
plt.plot(sAguacate, label="Bolsas S", color="black")
plt.plot(lAguacate, label="Bolsas L", color="cyan")
plt.plot(xlAguacate, label="Bolsas XL", color="yellow")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# renombrar columnas de la tabla
                        # nombre original:NOMBRE_NUEVO,
results.rename(columns={languages:"Languages", salary:"Salary"}, inplace=True) #remplazar nuevos nombres en la tabla
results.head(5)
```

```python
# eliminar NAN

results.dropna(subset=["Salary", "Languages"], inplace=True) #eliminar NAN de dos columnas, remplazar en tabla origial
results["Salary"].head(25) # No hay NAN en los primero 25, y en todos
# de 83439 datos se pasó a 46693
```

```python
results=results.loc[:,["Country", "Salary", "Languages"]].sort_values(by="Salary")
results #solo ver country, salary and languages, ordenar por salary

# hay dato atípicos, eliminarlos para un mejor análisis
```

```python
filtro= (results["Salary"] >=10000) & (results["Salary"] <= 3e6) #realizar filtro para eliminar datos atípicos

#reasignar resutlado a dato orignal:
results=results[filtro]
results
```

```python
# Separar cada lenguaje de programación y ponerle sus frecencias

lg=results["Languages"] #crear una varibale con los lenguajes
allLanguages=lg[21403].split(";") #separa los lenguajes, el registro 21403 es el que tiene más lenguajes, 
    #podría ser él el inicio para encontrar todos los lenguajes en la base
    
for row in lg: #ir de columna en columan tomando la columna de lg=Lenguages
    for lang in row.split(";"): # ir separando cada string de la columna Languages
        allLanguages.append(lang) # añadir cada string en la lista "lang"
        
allLanguages=list(set(allLanguages)) # hacer que la lista que guardó los lenguajes sólo se quede con los que NO
        # se repiten
# set es un tipo de dato, como la tupla, que elimina datos repetidos        
```

```python
allLanguages.sort()
allLanguages
```

```python
# crear un data FRAME donde los lenguajes sean los índices, y poner las medias de salarios en otra columnas
# import numpy as np

meanSalary=pd.DataFrame(data=np.zeros(len(allLanguages)), index=allLanguages, columns=["Salary"])
meanSalary # es un array con cero, para luego llenarlos
```
