# Creaci√≥n de Diargamas de Caja (BoxPlots)

```python
# importaci√≥n de librer√≠as:

import numpy as np
import matplotlib.pyplot as plt
```

```python
# convertir columnas del dataframe a arrays:

a_crec=np.array(tesis["crec_13_p"]) #convert√≠ col "cre_13_p" del DataFrame en un array
a_crec
```

```python
plt.yticks(np.arange(-15,15,2))#plano que inicie desde -15 hasta 15, de 2 en dos
plt.grid(True)#poner rayas al plano
plt.boxplot(a_crec)#grafica boxplot
plt.show()
```

```python
(np.min(a_crec), np.quantile(a_crec, 0.25), np.median(a_crec), #valor m√≠nimo; 1er cuartil; mediana;
 np.quantile(a_crec, 0.75), np.max(a_crec)) # tecer cuartil; valor m√°ximo
```

# Personalizaci√≥n de gr√°ficas

```python
plt.plot(tesis["t"], tesis["ar1"], label="nivel de RN1", color="purple", marker="<", linestyle="-")
plt.plot(tesis["t"], tesis["ar2"], label="nivel de RN2", color="red", marker="*", linestyle="--")
plt.plot(tesis["t"], tesis["ar3"], label="nivel de RN3", color="blue", marker="^", linestyle=":")
plt.plot(tesis["t"], tesis["ar4"], label="nivel de RN4", color="black", marker=".", linestyle="-.")
plt.plot(tesis["t"], tesis["crec_13_p"], label="crecimiento", color="green", marker="+", linestyle=" ")

#activar leyendas
plt.legend()

#t√≠tulo de gr√°fica:
plt.title("Nivel de recursos naturales y crecimiento")

#Etiquetas de eje:
plt.xlabel("A√±o de recursos/crecimiento")
plt.ylabel("Nil crecimiento")

#personalizar Eje de crecimiento de -15 a 15 en incremento de 3
plt.yticks(np.arange(-15,15,3))

#Activar cuadr√≠cula:
plt.grid()

#Activar marcas menores:
plt.minorticks_on()

plt.show()
```

# Matriz de correlaci√≥n

```python
matriz = tesis.corr() # -1 (existe una relaci√≥n fuerte)  0   1 (existe relaci√≥n fuerte)
plt.matshow(matriz, cmap="bwr", vmin=-1, vmax=1)
plt.xticks(range(14), tesis.columns, rotation=90)
plt.yticks(range(14), tesis.columns)

for i in range(len(matriz.columns)):
    for j in range(len(matriz.columns)):
        plt.text(i, j, round(matriz.iloc[i, j], 2),
                 ha="center", va="center")


plt.colorbar()
plt.show()
```

# Diagramas de dispersi√≥n

```python
print(">>>>>>>>>>>>>>>>>>>> SIMPLE")

sns.relplot(data=tesis,
           x="ar1",
           y="crec_13_p")

print(">>>>>>>>>>>>>>>>>>>> inseguridad Y crecimeinto")

fig = sns.relplot(data=tesis,
           x="i_inseguridad",
           y="pib2013_p",
           hue="i_inseguridad",
           size="i_inseguridad",
           sizes=(10, 100))

fig.savefig("out.png", transparent=True)


sns.relplot(data=tesis,
           x="ar1",
           y="crec_13_p",
           hue="abortolibre",
           s=300)

print(">>>>>>>>>>>>>>>>>>>> REJILLA, TAMA√ëO PERSONALIZADO Y ESTILO")

sns.relplot(data=tesis,
           x="ar1",
           y="crec_13_p",
           hue="t",
           row="abortolibre",
           size="i_inseguridad",
           sizes=(10, 100),
           style="abortolibre")

sns.relplot(data=tesis,
           x="ar1",
           y="crec_13_p",
           hue="t",
           col="entidad",
           row="abortolibre",
           size="i_inseguridad",
           sizes=(10, 100),
           style="abortolibre")
```
