# Regresión lineal

La regresión lineal es una técnica paramétrica de machine learning. Donde "paramétrica" hace referencia a que incluso antede de mirar a los datos, se sabe cuántos parámetros (o coeficietnes) vamos a necesitar.


En el caso que estemos usando una sola variable, x, sabemos que una línea necesita 2 parámetros. La fórmula para la regresión lineal con una variable x es:

y = wx + b


El aprendizaje consiste en encontrar cuáles son los mejores parámetros (coeficentes) para los datos que tengamos. Los mejores coeficientes serán los que minimicen alguna medida de error. Para la regresión lineal usaremos el error cuadrático medio.

image.png

Al usar una regresión lineal para encontrar los parámetros de la línea que miniza el error de los datos que tenemos. El proceso de aprendizaje consiste en estimar los parámetros w y b.

w=0.0918
b=1.2859

así nos queda:
y = 0.0918x + 1.2859

Podemos usar este modeo de regresión lineal para estimar cuáles serán los resultados para otros valores de x. Por ejemplo, si queremos saber el resultado para x=5, usaremos el modelo anterior y veremos que el resultado es 1.7449:

y = 0.0918(5) + 1.2859 = 1.7449

# objetivo de Regresión linea

El objetivo es minimizar la distancia vertical entre todos los datos y nuestra línea que vendría siendo el modelo; para determinar la mejor línea debemos mimimizar la distancia entre todos los puntos y la distancia de nuestra línea (por ejemplo, con el método de mímimos cuadrados).
image.png

# cálculo del Error Cuadrático Medio

Una regresión lineal (línea azul) que estima los datos (puntos verdes) presenta errores (en rojo) que se define con la siguiente fórmula:

error_cuadrativo = (real - estimado)^2

El valor estimado es el que nos da el modelo (color azul); el error perfecto es 0.
Al calcular el error en cada punto es posible calcular cual es el error medio. Para ello, se debe sumar todos los errores y los dividimos entre el número total de puntos. Si llamamos M al número total de puntos nos queda la fórmula del Error Cuadrática Medio (MSE, Mean Squared Error):

image.png

Otras formas de resolver el problema de regresión linea es utilizando métodos de optimización numérica tales como el gradiente descendiente.

El gradiente descendiente va a servir no sólo para resolver regresión lineales y polinómias sino que es también para el aprendizaje automático de redes neuronales y aprendizaje profundo.

# Raíz Cuadrada del Error Cuadratico Medio

El error medio cuadrático medio no es del todo intuitivo porque nos da el error medio al cuadrado. Si nuestro modelo predice el valor de un BIEN INMUEBLE presenta un error medio duadrático de 1,000,000 unidades monetarias nos daría la impresión de que es error es mucho. **texto en negrita** En realidad el error sería más/menos 1,000 unidades monetarias en mida, porque la raíz cuadrada de 1,000,000 es 1,000. f´romula:

image.png

EL RMSE (Root Mean Squared Error) es más intuitivo que el MSE, pero no se usa porque computacionalmente es más costoso (calcular el error en cada iteración es lento), y el MSE permite que algoritmos de optimización, como el gradiente descendiente, encuentre una solución más rápido.
Normalmente se suele usar el MSE durante el proceso de aprendizaje y su raíz cuadrada al final, para dar una estimación en términos intuitivos de la calidad de la predicción.

# Regresión lineal

```python

# Imports necesarios
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
bd.shape
```

```python

# Visualizamos rápidamente las caraterísticas de entrada
bd.drop(['ar4', "ar2", "ar3",'abortolibre', 'i_corrupcion', "pib2013_p", "crec_13_p",
         "t"],1).hist()
plt.show()
```

```python
dataX =bd[["ar1"]] # es la  variable independiente/explicariva/variable de entrada:x:ar1
X_train = np.array(dataX) #convertir vairiable de entrada en un array
y_train = bd['pib2013_p'].values #variable dependiente:PIB

# Creamos el objeto de Regresión Linear
regr = linear_model.LinearRegression()

# Entrenamos nuestro modelo
regr.fit(X_train, y_train)

# Hacemos las predicciones que en definitiva una línea (en este caso, al ser 2D)
y_pred = regr.predict(X_train) #PIB que se predice con regresión

# Veamos los coeficienetes obtenidos, En nuestro caso, serán la Tangente
print('Coefficients: \n', regr.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('Independent term: \n', regr.intercept_)
# Error Cuadrado Medio
print("Mean squared error:", mean_squared_error(y_train, y_pred))
# Puntaje de Varianza. El mejor puntaje es un 1.0
print('Variance score:', r2_score(y_train, y_pred))
```

# Predicción en regresión lineal simple

```python
# con nuestro modelo vamos a predecir cuál será el PIB real si ar1 = 0.11

# según nuestro modelo, hacemos:
y_ar_011 = regr.predict([[0.11]])
print("El PIB real promedio de la entidade será de: {} cuando ar1 sea: {}".format(int(y_ar_011), 0.11) )
```

# Regresión lineal múltiple

ecuación sera:

Y = b + m1 X1 + m2 X2 + … + m(n) X(n)

〖crec_13_p〗_it=α+β_1 〖ar1〗_it+institucion_media_iycx_2it+ u_it

```python
# Para poder graficar en 3D, haremos una variable nueva
media= (bd["i_inseguridad"].fillna(0) + bd['i_corrupcion'].fillna(0))

dataX2 =  pd.DataFrame() #creamo un DataFrame
dataX2["ar1"] = bd["ar1"] #agregar ar1 en el anterior DataFrame
dataX2["media"] = media #agregar esta variable en el DataFrame
XY_train = np.array(dataX2) #convertir las variables anterio en un array <--varoables de entrada
z_train = bd['pib2013_p'].values #PIB real, valor para entrar modelo <--variable de salida
```

```python
# crear un nuevo objeto de regresión lineal pero con dos dimensiones que entrenar: XY_train.


regr2 = linear_model.LinearRegression()

# Entrenamos el modelo, esta vez, con 2 dimensiones
# obtendremos 2 coeficientes, para graficar un plano
regr2.fit(XY_train, z_train)

# Hacemos la predicción con la que tendremos puntos sobre el plano hallado
z_pred = regr2.predict(XY_train) #valor de PIB predecido con modelo

# Los coeficientes
print('Coefficients: \n', regr2.coef_)
# Error cuadrático medio
print("Mean squared error: {}".format(mean_squared_error(z_train, z_pred)))
# Evaluamos el puntaje de varianza (siendo 1.0 el mejor posible)
print('Variance score: {}'.format(r2_score(z_train, z_pred)))
```

```python
fig = plt.figure()
ax = Axes3D(fig)

# Creamos una malla, sobre la cual graficaremos el plano
xx, yy = np.meshgrid(np.linspace(0, 1, num=1), np.linspace(0, 3, num=2))

# calculamos los valores del plano para los puntos x e y
nuevoX = (regr2.coef_[0] * xx)
nuevoY = (regr2.coef_[1] * yy)

# calculamos los correspondientes valores para z. Debemos sumar el punto de intercepción
z = (nuevoX + nuevoY + regr2.intercept_)

# Graficamos el plano
ax.plot_surface(xx, yy, z, alpha=0.2, cmap='hot')

# Graficamos en azul los puntos en 3D
ax.scatter(XY_train[:, 0], XY_train[:, 1], z_train, c='blue',s=3)

# Graficamos en rojo, los puntos que
ax.scatter(XY_train[:, 0], XY_train[:, 1], z_pred, c='red',s=4)

# con esto situamos la "camara" con la que visualizamos
ax.view_init(elev=25., azim=225)

ax.set_xlabel('Indice de recursos naturales')
ax.set_ylabel('Suma de índice de inseguridad y corrupción')
ax.set_zlabel('????')
ax.set_title('Regresión Lineal con Múltiples Variables')
```
